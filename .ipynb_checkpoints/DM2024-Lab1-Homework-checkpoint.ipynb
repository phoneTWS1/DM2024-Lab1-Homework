### 3.5 Finding frequent patterns

First, create separate DataFrames for each sentiemnt

````python
sentiments = ['not nostalgia','nostalgia']
sentiment_dfs = {}
for sentiment in sentiments:
    sentiment_dfs[sentiment] = X[X['sentiment'] == sentiment].copy()

sentiment_dfs['not nostalgia'].head(3)
````

Create term-document frequency DataFrame for each sentiment

````python
# Function to create term-document frequency DataFrame for each sentiment
def create_term_document_df(df):
    count_vect = CountVectorizer()
    X_counts = count_vect.fit_transform(df['comment'])
    words = count_vect.get_feature_names_out()
    term_document_df = pd.DataFrame(X_counts.toarray(), columns=words)
    return term_document_df

# Create term-document frequency DataFrames for each category
filt_term_document_dfs = {}
for sentiment in sentiments:
    filt_term_document_dfs[sentiment] = create_term_document_df(sentiment_dfs[sentiment])

filt_term_document_dfs['not nostalgia'].head(3)
````

In the previous sections, we examined the frequency of each word in the documents. However, we want to generalize further to observe and determine the data we will use to mine patterns. To achieve this, we will group the terms into bins and plot their frequencies.

````python
# Function to plot the distribution of frequency
def plot_term_freq_distribution(sentiment_num,dfs,bins):
    sentiments = ['not nostalgia','nostalgia']
    word_counts = dfs[sentiments[sentiment_num]].sum(axis=0).to_numpy()

    plt.hist(word_counts, bins=bins,color='blue')
    plt.title(f'Term Frequency Distribution for Sentiment {sentiments[sentiment_num]}')
    plt.xlabel('Frequency')
    plt.ylabel('Number of Terms')
    #plt.xlim(0, 200)

plt.figure(figsize=(15, 3))
plt.subplot(1,2,1)
plot_term_freq_distribution(0,filt_term_document_dfs,150)
plt.subplot(1,2,2)
plot_term_freq_distribution(1,filt_term_document_dfs,150)
````

We observe that most words appear infrequently across the entire dataset, while a small number of words are quite common.

To enhance our results, we will filter out these words. Specifically, we'll remove the bottom 75% of the least frequent words and the top 1.7% of the most frequent words in the 'not nostalgia' dataset, as well as the bottom 75% of the least frequent words and the top 2.5% of the most frequent words in the 'nostalgia' dataset. This approach helps us focus on words that may reveal more valuable patterns."

````python

# Filter the bottom 1% and top 5% words based on their sum across all documents
def filter_top_bottom_words_by_sum(term_document_df, top_percent=0.05, bottom_percent=0.01):
    # Calculate the sum of each word across all documents
    word_sums = term_document_df.sum(axis=0)
    
    # Sort the words by their total sum
    sorted_words = word_sums.sort_values()
    
    # Calculate the number of words to remove
    total_words = len(sorted_words)
    top_n = int(top_percent * total_words)
    bottom_n = int(bottom_percent * total_words)
    
    # Get the words to remove from the top 5% and bottom 1%
    words_to_remove = pd.concat([sorted_words.head(bottom_n), sorted_words.tail(top_n)]).index
    print(f'Bottom {bottom_percent*100}% words: \n{sorted_words.head(bottom_n)}') #Here we print which words correspond to the bottom percentage we filter
    print(f'Top {top_percent*100}% words: \n{sorted_words.tail(top_n)}') #Here we print which words correspond to the top percentage we filter
    # Return the DataFrame without the filtered words
    print('\n')
    return term_document_df.drop(columns=words_to_remove)

# Apply the filtering function to each category
term_document_dfs = {}
term_document_dfs['not nostalgia'] = filter_top_bottom_words_by_sum(filt_term_document_dfs['not nostalgia'],0.017,0.75)
term_document_dfs['nostalgia'] = filter_top_bottom_words_by_sum(filt_term_document_dfs['nostalgia'],0.025,0.75)

plt.figure(figsize=(15, 3))
plt.subplot(1,2,1)
plot_term_freq_distribution(0,term_document_dfs,100)
plt.subplot(1,2,2)
plot_term_freq_distribution(1,term_document_dfs,100)

print('orignal term document shape of not nostalgia', filt_term_document_dfs['not nostalgia'].shape)
print('filtered term document shape of not nostalgia', term_document_dfs['not nostalgia'].shape)
print('orignal term document shape of nostalgia', filt_term_document_dfs['nostalgia'].shape)
print('filtered term document shape of nostalgia', term_document_dfs['nostalgia'].shape)
````

After filtering out overly common and rare words, the figure above shows that the distribution of frequencies is now more balanced.

 Identify the unique patterns that are specific to each sentiment.

````python
dfs = [frequentPatternsDF_not_nostalgia, frequentPatternsDF_nostalgia]

pattern_counts = {}
for df in dfs:
    for pattern in df['Patterns']:
        if pattern not in pattern_counts:
            pattern_counts[pattern] = 1
        else:
            pattern_counts[pattern] += 1

unique_patterns = {pattern for pattern, count in pattern_counts.items() if count==1}
filtered_dfs = []
for df in dfs:
    filtered_df = df[df['Patterns'].isin(unique_patterns)]
    filtered_dfs.append(filtered_df)

final_pattern_df = pd.concat(filtered_dfs, ignore_index=True)

final_pattern_df = final_pattern_df.sort_values(by='Support', ascending=False)

final_pattern_df
````

Integrate these now sentiment-specific patterns into our data.

````python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Convert 'text' column into term-document matrix using CountVectorizer
count_vect = CountVectorizer()
X_tdm = count_vect.fit_transform(X['comment'])  # X['text'] contains your text data
terms = count_vect.get_feature_names_out()  # Original terms in the vocabulary

# Tokenize the sentences into sets of unique words
X['tokenized_comment'] = X['comment'].str.split().apply(set)

# Initialize the pattern matrix
pattern_matrix = pd.DataFrame(0, index=X.index, columns=final_pattern_df['Patterns'])

# Iterate over each pattern and check if all words in the pattern are present in the tokenized sentence
for pattern in final_pattern_df['Patterns']:
    pattern_words = set(pattern.split())  # Tokenize pattern into words
    pattern_matrix[pattern] = X['tokenized_comment'].apply(lambda x: 1 if pattern_words.issubset(x) else 0)
pattern_matrix.head(3)
````

### 3.6.Dimensionality Reduction

Apply  Principal Component Analysis (PCA), t-Distributed Stochastic Neighbor Embedding (t-SNE), Uniform Manifold Approximation and Projection (UMAP), to find projections that captures the largest amount of variation in the data. 

````python
#Applying dimensionality reduction with only the document-term frequency data
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import matplotlib.pyplot as plt
````

````python
# Plot the results in subplots
col = ['blue', 'orange']
sentiments = X['sentiment'].unique() 

# Define a function to create a scatter plot for each method
def plot_scatter(ax, X_reduced, title):
    for c, sentiment in zip(col, sentiments):
        xs = X_reduced[X['sentiment'] == sentiment].T[0]
        ys = X_reduced[X['sentiment'] == sentiment].T[1]
        ax.scatter(xs, ys, c=c, marker='o', label=sentiment)
    
    ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend(loc='upper right')
````

The above figure show that both t-SNE and UMAP can give a slight hint of separation in the data.

 ### Naive Bayes classifier train with TF-IDF features

````python
def select_feature_with_tfidf(df,top_n):
    count_vect = CountVectorizer()
    df_counts = count_vect.fit_transform(df['comment'])
    words = count_vect.get_feature_names_out()
    tdm_df = pd.DataFrame(df_counts.toarray(), columns=words)
    
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(df['comment'])
    terms = vectorizer.get_feature_names_out()
    terms_tfidf = tfidf_matrix.toarray().sum(axis=0)
    terms_tfidf_df = pd.DataFrame({'term': terms, 'tfidf':terms_tfidf})
    tfidf_sort = term_tfidf_df.sort_values(by='tfidf', ascending=False)
    select_term_df = tdm_df[list(tfidf_sort['term'][:top_n])]

    return select_term_df
````

To determine the optimal number of terms (top_n) for training the classifier, we conducted an experiment that iterates over 80 different values of top_n, comparing their accuracy, precision, and recall."

````python
top_n = list(range(10,2000,25))
accuracy = []
precision = []
recall = []
class_reports = []
nb_classifier = {}
target_names = sentiments
for t in top_n:
    select_term_df = select_feature_with_tfidf(X,t)
    X_train,X_test,y_train,y_test = train_test_split(select_term_df, X['sentiment'], test_size=0.3, random_state=42)
    nb_classifier[t] = MultinomialNB()
    nb_classifier[t].fit(X_train, y_train)
    y_pred = nb_classifier[t].predict(X_test)
    accuracy.append(accuracy_score(y_test,y_pred))
    precision.append(precision_score(y_test,y_pred,pos_label='not nostalgia'))
    recall.append(recall_score(y_test,y_pred,pos_label='not nostalgia'))
    class_reports.append(classification_report(y_test, y_pred, target_names=target_names, digits=2))
````

````python
plt.plot(top_n, accuracy, label='accuracy')
plt.plot(top_n, precision, label='precision')
plt.plot(top_n, recall, label='recall')
plt.xlabel('number of feature used')
plt.ylabel('accuracy')
plt.legend()
plt.show()
````

Suprisingly, the result shows that near 250 term can give the best performance. 

````python
opt_index = accuracy.index(max(accuracy))
top_n[opt_index]
print('train with top {:d} feature'.format(top_n[opt_index]))
print("Accuracy:", accuracy[opt_index])
print("\nClassification Report:\n", class_reports[opt_index])
````

We compared a Naive Bayes classifier trained on all word frequencies with another Naive Bayes classifier trained on the top 260 terms selected by TF-IDF values. The result show that the latter one can perform better. This is surprising, considering the significant difference in the number of features used to train these classifiers. This implies that TF-IDF can be very effective for feature selection, and that the selection of features can significantly impact the performance of a classifier.
